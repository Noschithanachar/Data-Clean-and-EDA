```python
df['OccurrenceDate'] = pd.to_datetime(df['OccurrenceDate'], errors='coerce')
df['ReportDate'] = pd.to_datetime(df['ReportDate'], errors='coerce')
df['ClosingDate'] = pd.to_datetime(df['ClosingDate'], errors='coerce')
```

```python
df['year'] = df['OccurrenceDate'].dt.year
df['month'] = df['OccurrenceDate'].dt.month
```

```python
import matplotlib.pyplot as plt

df['year'].value_counts().sort_index().plot(kind='bar')
plt.title("Incidents per Year")
plt.show()
```

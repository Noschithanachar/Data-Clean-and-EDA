```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("D:\\EDA\\eda\\rawdata\\Single_view_of_ship.csv")
```

```python
data
```

```python
data.columns
```

```python
data.shape
data.info()
data.describe()
data.head()
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')
```

```python
data['first_event_of_ship'] = pd.to_datetime(data['first_event_of_ship'], errors='coerce')
data['last_event_of_ship'] = pd.to_datetime(data['last_event_of_ship'], errors='coerce')
```

```python
threshold = len(data) * 0.7
data = data.dropna(thresh=threshold, axis=1)
```

```python
data.fillna(0, inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data['incident_rate'] = data['ninc'] / data['life_time_of_ship_in_months']
data['near_miss_rate'] = data['nnm_total'] / data['life_time_of_ship_in_months']
data['adi_rate'] = data['nadi_total'] / data['life_time_of_ship_in_months']

data['year'] = data['first_event_of_ship'].dt.year
```

```python
data[['ninc','nnm_total','nadi_total']].hist(figsize=(10,5))
```

```python
numeric_data = data.select_dtypes(include=['number'])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_data.corr(), cmap='coolwarm')
```

```python
data_corr = data.drop(columns=['name_of_ship', 'first_event_of_ship', 'last_event_of_ship'])

sns.heatmap(data_corr.corr(), cmap='coolwarm')
```

```python
sns.heatmap(data.corr(numeric_only=True), cmap='coolwarm')
```

```python
corr = numeric_data.corr()

# Select top correlated features with ninc
top_corr = corr['ninc'].sort_values(ascending=False).head(10)

top_cols = top_corr.index

sns.heatmap(numeric_data[top_cols].corr(), annot=True, cmap='coolwarm')
```
